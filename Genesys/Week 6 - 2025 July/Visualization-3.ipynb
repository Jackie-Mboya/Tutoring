### Import Python Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#years=range(2018, 2023)
```

```python
## Random data points for yield of apples over the years
years=[2018, 2019, 2020, 2021, 2022, 2023]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931] ## yields of apples in tonnes per hectare
```

### Using Matplotlib

```python
#plotting data points
plt.plot(yield_apples)
```

```python
#add x-axis labels to understand the data and its purpose
plt.plot(years, yield_apples)
```

```python
#Add labels
plt.plot(years, yield_apples)
plt.xlabel("Years")
plt.ylabel("Yield of Apples (in tons)")
```

```python
yield_oranges=[0.9, 0.92, 0.95, 0.96, 0.97, 0.98]
```

```python
#plot multiple
plt.plot(years, yield_apples)
plt.plot(years, yield_oranges)
plt.xlabel("Years")
plt.ylabel("Yield (in tons)")
```

```python
## Add legends
plt.plot(years, yield_apples)
plt.plot(years, yield_oranges)
plt.xlabel("Years")
plt.ylabel("Yield (in tons)")
plt.legend(["Apples", "Oranges"])
plt.show()
```

```python
#add markers to the plot
plt.plot(years, yield_apples, marker='o')
plt.plot(years, yield_oranges, marker='x')
plt.xlabel("Years")
plt.ylabel("Yield (in tons)")
plt.legend(["Apples", "Oranges"])
plt.show()
```

## Using Seaborn

```python
#set global seaborn style
sns.set_style("darkgrid")
```

```python
# Plotting the data
plt.figure(figsize=(10, 6)) # set the figure size to change the size
plt.plot(years, yield_apples, marker='o', label='Apples') #plot apples
plt.plot(years, yield_oranges, marker='x', label='Oranges') #plot oranges
plt.xlabel("Years") # add x-axis label
plt.ylabel("Yield (in tons)") # add y-axis label    
plt.title("Fruit Yield Over Years") # add title to the plot
plt.legend() # add legend to the plot
plt.show() # display the plot
```

### Bar Graphs

```python
plt.bar(years, yield_apples, label='Apples', color='blue', alpha=0.6)
plt.xlabel("Years") # add x-axis label
plt.ylabel("Yield (in tons)") # add y-axis label   
plt.title("Fruit Yield Over Years") # add title to the plot
#plt.legend() # add legend to the plot
plt.show() # display the plot
```

```python
#stacked bar graph
plt.bar(years, yield_apples, label='Apples', color='blue', alpha=0.6)
plt.bar(years, yield_oranges, bottom=yield_apples, label='Oranges', color='orange', alpha=0.6)
plt.xlabel("Years") # add x-axis label
plt.ylabel("Yield (in tons)") # add y-axis label
plt.title("Fruit Yield Over Years") # add title to the plot
plt.legend() # add legend to the plot
plt.show() # display the plot
```

```python
#stacked bar graph
plt.bar(years, yield_apples, label='Apples', bottom=yield_oranges, color='blue', alpha=0.6)
plt.bar(years, yield_oranges, label='Oranges', color='orange', alpha=0.6)
plt.xlabel("Years") # add x-axis label
plt.ylabel("Yield (in tons)") # add y-axis label
plt.title("Fruit Yield Over Years") # add title to the plot
plt.legend() # add legend to the plot
plt.show() # display the plot
```
