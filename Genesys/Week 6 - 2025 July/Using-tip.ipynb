### Presenting Data Insights to Stakeholders - Data Storytelling
- Introduction: State the problem/question.
- Methodology (brief): How did you get the data?
- Key Findings (with visuals): Present most important insights.
- Implications/Recommendations: What do findings mean? What actions?
- Conclusion & Q&A: Summarize and invite questions

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load in-built tips dataset from seaborn
tips_data = sns.load_dataset("tips")
tips_data
```

### Scatter plot

```python
# Scatter plot with day against tip
plt.scatter(tips_data['day'], tips_data['tip'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

```python
## Add colors and change the size of the points
plt.scatter(tips_data['day'], tips_data['tip'], c=tips_data["size"], s=tips_data['total_bill'], alpha=0.5)
```

## Bar Chart

```python
#comparison, side-by-side
sns.barplot(data=tips_data, x="day", y="total_bill", hue="sex")
```
