```python
#Load Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a sample DataFrame
data = {
    'Month': ['Jan', 'Feb', 'March', 'April', 'May', 'June'],
    'Sales': [150, 180, 130, 250, 150, 100]
}
df = pd.DataFrame(data)
```

```python
# Plotting the data
# --- Bar Chart for Sales Trend ---
plt.figure(figsize=(10, 6)) # Set figure size for better readability
sns.barplot(x='Month', y='Sales', data=df, palette='viridis') # Using seaborn for a nice bar plot
plt.title('Monthly Sales Performance', fontsize=16) # Add a clear title
plt.xlabel('Month', fontsize=12) # Label the x-axis
plt.ylabel('Sales', fontsize=12) # Label the y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7) # Add horizontal grid lines for easier reading
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show() # Display the plot
```

```python
# --- Line Graph for Sales Trend ---
plt.figure(figsize=(10, 6)) # Set figure size
sns.lineplot(x='Month', y='Sales', data=df, marker='o', color='red', linewidth=2) # Using seaborn for line plot with markers
plt.title('Monthly Sales Trend', fontsize=16) # Add a clear title
plt.xlabel('Month', fontsize=12) # Label the x-axis
plt.ylabel('Sales', fontsize=12) # Label the y-axis
plt.grid(True, linestyle='--', alpha=0.7) # Add grid lines
plt.ylim(0, max(df['Sales']) * 1.1) # Set y-axis limit to start from 0 and have some padding
plt.tight_layout() # Adjust layout
plt.show() # Display the plot
```

```python
# --- Pie Chart for Sales Distribution ---
plt.figure(figsize=(8, 8)) # Set figure size for pie chart
plt.pie(df['Sales'], labels=df['Month'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel')) # Create pie chart with percentages
plt.title('Sales Distribution by Month', fontsize=16) # Add a title 
plt.axis('equal') # Equal aspect ratio ensures that pie chart is a circle
plt.show() # Display the pie chart
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# --- Sample Data for Demonstrations ---
np.random.seed(42) # for reproducibility

# Data for Comparison (Bar, Line, Scatter)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_data = {
    'Month': months,
    'Sales_A': np.random.randint(100, 300, 12),
    'Sales_B': np.random.randint(80, 280, 12)
}
df_sales = pd.DataFrame(sales_data)
```

```python
# Data for Distribution (Histograms, Box plots, Density plots)
data_dist = {
    'Value': np.random.normal(loc=50, scale=10, size=1000),
    'Category': np.random.choice(['Group 1', 'Group 2', 'Group 3'], size=1000)
}
df_dist = pd.DataFrame(data_dist)
```

```python
# Data for Composition (Stacked Bar, Treemap)
composition_data = {
    'Product_Category': ['Electronics', 'Clothing', 'Home Goods', 'Books'],
    'Q1_Sales': [250, 180, 120, 70],
    'Q2_Sales': [280, 200, 150, 90],
    'Q3_Sales': [300, 220, 130, 80]
}
df_composition = pd.DataFrame(composition_data)
```

```python
# Data for Relationship (Scatter, Heatmap)
data_rel = {
    'Feature_X': np.random.rand(50) * 10,
    'Feature_Y': np.random.rand(50) * 10 + (np.random.rand(50) * 0.8 * np.random.rand(50) * 10), # Some correlation
    'Feature_Z': np.random.rand(50) * 5
}
df_rel = pd.DataFrame(data_rel)
```

```python
# Data for Heatmap
corr_matrix = df_rel.corr(numeric_only=True)

print("--- Comparison Charts ---")
```

```python
# 1. Bar Chart (Comparison)
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Sales_A', data=df_sales, palette='Blues_d')
plt.title('Monthly Sales Comparison (Sales_A)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2. Line Chart (Comparison / Trend over Time)
plt.figure(figsize=(12, 7))
sns.lineplot(x='Month', y='Sales_A', data=df_sales, marker='o', label='Sales A', color='blue')
sns.lineplot(x='Month', y='Sales_B', data=df_sales, marker='x', label='Sales B', color='green', linestyle='--')
plt.title('Monthly Sales Trend Comparison')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 3. Scatter Plot (Comparison / Relationship)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sales_A', y='Sales_B', data=df_sales, hue='Month', size='Sales_A', sizes=(20, 400), palette='viridis', alpha=0.8)
plt.title('Sales A vs. Sales B by Month')
plt.xlabel('Sales A')
plt.ylabel('Sales B')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
print("\n--- Distribution Charts ---")
```

```python
# 4. Histogram (Distribution)
plt.figure(figsize=(9, 6))
sns.histplot(df_dist['Value'], bins=20, kde=True, color='purple', edgecolor='black')
plt.title('Distribution of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 5. Box Plot (Distribution)
plt.figure(figsize=(9, 6))
sns.boxplot(x='Category', y='Value', data=df_dist, palette='pastel')
plt.title('Value Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```

```python
# 6. Density Plot (KDE Plot - Distribution)
plt.figure(figsize=(9, 6))
sns.kdeplot(x='Value', hue='Category', data=df_dist, fill=True, palette='coolwarm', common_norm=False)
plt.title('Density Plot of Values by Category')
plt.xlabel('Value')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
print("\n--- Composition Charts ---")

# 7. Stacked Bar Chart (Composition)
df_composition_melted = df_composition.melt(id_vars='Product_Category', var_name='Quarter', value_name='Sales')
plt.figure(figsize=(10, 7))
sns.barplot(x='Product_Category', y='Sales', hue='Quarter', data=df_composition_melted, palette='magma')
plt.title('Quarterly Sales Composition by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(title='Quarter')
plt.tight_layout()
plt.show()
```

```python
# 8. Treemap (Composition - using Plotly for interactivity)
# Note: Treemaps are not natively in Matplotlib/Seaborn, Plotly is a good choice.
print("Generating interactive Treemap (opens in a new window/tab if run locally)...")
fig_treemap = px.treemap(df_composition_melted, path=[px.Constant("All Products"), 'Product_Category', 'Quarter'], values='Sales',
                         title='Sales Composition by Product Category and Quarter (Interactive Treemap)')
fig_treemap.show()
print("Treemap generated.")
```

```python
print("\n--- Relationship Charts ---")

# 9. Scatter Plot (Relationship) - already covered in Comparison, but re-emphasize for relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Feature_X', y='Feature_Y', data=df_rel, hue='Feature_Z', size='Feature_Z', sizes=(50, 800), palette='plasma', alpha=0.7)
plt.title('Relationship between Feature X and Feature Y (colored by Feature Z)')
plt.xlabel('Feature X')
plt.ylabel('Feature Y')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python


# 10. Heatmap (Relationship - Correlation Matrix)
plt.figure(figsize=(8, 7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Features')
plt.tight_layout()
plt.show()

print("\nAll chart examples generated successfully.")
```
