## Data Visualization - Week 6

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create a sample DataFrame
data = {
    "Fruits":["Orange", "Apple", "Banana", "Grapes", "Mango", "Pineapple"],
    "Quantity":[10, 15, 7, 20, 12, 5],
    "Price":[1.5, 2.0, 0.5, 3.0, 2.5, 4.0]
}
```

```python
df = pd.DataFrame(data)
df.head(10)
```

```python
df['Total'] = df['Quantity'] * df['Price']
df
```

```python
df['Currency'] = 'USD'
df.head()
```

```python
#Set color
colors = plt.cm.viridis(range(len(df)))
```

```python
#Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Fruits', y='Quantity', palette="viridis")
plt.title("Quantity of Fruits Sold")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.xticks(rotation=45)
plt.show()
```

```python
#Heatmap of Fruits
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Quantity', 'Price', 'Total']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Fruits Data")
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time", palette="deep")
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip")
plt.title("Bill vs Tip")
plt.show()
```
