```python
#While loops
x = 2
while x < 20:
    print(x)
    x += 1
```

```python
#While loops
x = 2
while x < 20:
    print(x)
    x += 1
```

```python
#While loops
x = 2
while x < 20:
    print(x)
    if x==10:
        break
    x += 1
```

```python
#While loops continue
x = 2
while x < 20:
    x += 1
    if x==9:
        continue
    print(x)
```

```python
# while ... else loop 
x = 2
while x < 20:
  print(x)
  x += 1
else:
  print("x is no longer less than 20")
```

```python
import matplotlib.pyplot as plt

# Create a dictionary with keys and values
mydict = {"A": 20, "B": 30, "C": 50, "D": 60, "E": 20, "F": 30, "G": 100}

# Extract keys and values for plotting
keys = list(mydict.keys())
values = list(mydict.values())

# Plot the data
plt.bar(keys, values, color="skyblue")

# Add labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart of My Dictionary Data")

# Show the plot
plt.show()
```
